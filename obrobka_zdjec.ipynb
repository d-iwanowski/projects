```python
for i, file in enumerate(os.listdir(directory)):
    
    filename = os.fsdecode(file)
    img = cv2.imread(os.path.join(directory, filename))
    if len(img) < 1000:
        cropped_img = img[60:860, 60:1259] # photo horizontaly
    else:
        cropped_img = img[64:1263, 64:864] # photo verticaly
        
    j = str(i + 1)
    
    # Saving a file with file name like "photo###"
    newfilename = f"cropped_photos/photo{str(j).zfill(3)}.jpg"

    cv2.imwrite(newfilename, cropped_img)
```
